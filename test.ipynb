```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Donn√©es simples
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 5, 4, 5])

# Mod√®le
model = LinearRegression()
model.fit(X, y)

# Pr√©diction
X_test = np.array([[6]])
prediction = model.predict(X_test)

print(f"Pr√©diction pour 6 : {prediction[0]:.2f}")

# Affichage graphique
plt.scatter(X, y, color='blue', label='Donn√©es')
plt.plot(X, model.predict(X), color='red', label='Mod√®le')
plt.scatter(X_test, prediction, color='green', label='Pr√©diction pour 6')
plt.legend()
plt.title("R√©gression Lin√©aire")
plt.show()
```
